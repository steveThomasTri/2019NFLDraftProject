```python
import numpy as np
import pandas as pd
```

```python
combine = pd.read_csv("testcombine2019.csv")
prospects = pd.read_csv("prospectinfo.csv")
prospects.head()
```

```python
playerdata = combine.merge(prospects, how="left", on="player")
```

```python
playerdata.head()
```

```python
#colleges = pd.read_csv("collegescores.csv")
playerdata["college"] = playerdata["college"].replace({"BYU":"Brigham Young","TCU":"Texas Christian"})
#playerdata = playerdata.merge(colleges, how="left", on="college")
```

```python
#playerdata = playerdata.drop(["college","position"],1)
#playerdata.to_csv("testingcombine2019.csv", index=False)
playerdata = playerdata.rename(columns={"bench":"benchpress","vert":"vert_leap_in","broad":"broad_jump_in","cone":"3cone","twenty":"shuttle","forty":"40yards",'height':"height_in","weight":"weight_lbs"})
playerdata
```

```python
for index, row in playerdata.iterrows():
    height = row["height_in"].split("-")
    height = int(height[0]) * 12 + int(height[1])
    playerdata.loc[index, "height_in"] = height
```
