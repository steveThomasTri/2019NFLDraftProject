```python
from splinter import Browser
import bs4 as BeautifulSoup
import pandas as pd
import time
import numpy as np
```

```python
executable_path = {'executable_path':"chromedriver-windows.exe"}
page_link = "http://www.nfl.com/combine/tracker#day=fullresults"
browser = Browser('chrome',**executable_path, headless=False)
browser.driver.set_window_size(1360, 650)
browser.visit(page_link)
```

```python
prospectives = []
```

```python
nxtbutton = True
while(nxtbutton):
    rows = browser.find_by_css("div.tracker-table tbody tr")
    
    for row in rows:
        prospectives.append(row.find_by_css("td:nth-child(1)").text)

    if (len(browser.find_by_css("li.next a")) > 0):
        browser.find_by_css("li.next a").click()
    else:
        nxtbutton = False
        
prospect_df = pd.DataFrame({"player":prospectives})
        
```

```python
for event in ["bench","vert","broad","cone","twenty","forty"]:
    listing = []
    
    browser.reload()
    time.sleep(2)
    browser.find_by_css("a."+event).click()
    
    nxtbutton = True
    while(nxtbutton):
        rows = browser.find_by_css("div.tracker-table tbody tr")
        time.sleep(3)

        for roww in browser.find_by_css("div.tracker-table tbody tr"):
            listing.append(roww.find_by_css("td:nth-child(4)").text)

        if (len(browser.find_by_css("li.next a")) > 0):
            browser.find_by_css("li.next a").click()
        else:
            nxtbutton = False
            prospect_df[event] = listing
```

```python
prospect_df = prospect_df.replace("-",np.nan)
prospect_df
```

```python
prospect_df = prospect_df.apply(pd.to_numeric, errors="ignore")
prospect_df
```

```python
executable_path = {'executable_path':"chromedriver-windows.exe"}
page_link = "https://www.cbssports.com/nfl/draft/prospect-rankings/"
browser = Browser('chrome',**executable_path, headless=False)
browser.driver.set_window_size(1360, 650)
browser.visit(page_link)
```

```python
chroma = browser.find_by_xpath('//*[@id="video-playlist"]/section/div/div/div[3]/table/tbody/tr')
```

```python
prospects = []
colleges = []
positions = []
heights = []
weights = []
```

```python
for txt in chroma:
    try:
        if (txt.find_by_css("td:nth-child(2)").text.strip() != ""):
            name = txt.find_by_css("td:nth-child(2)").text
            name = name.split(" ")
            prospects.append(f"{name[1]}, {name[0]}")
    except:
        pass
    
    try:
        if (txt.find_by_css("td:nth-child(3)").text.strip() != ""):
            colleges.append(txt.find_by_css("td:nth-child(3)").text.strip())
    except:
        pass
    
    try:
        if (txt.find_by_css("td:nth-child(5)").text.strip() != ""):
            positions.append(txt.find_by_css("td:nth-child(5)").text.strip())
    except:
        pass
    
    try:
        if (txt.find_by_css("td:nth-child(7)").text.strip() != ""):
            heights.append(txt.find_by_css("td:nth-child(7)").text.strip())
    except:
        pass
    
    try:
        if (txt.find_by_css("td:nth-child(8)").text.strip() != ""):
            weights.append(txt.find_by_css("td:nth-child(8)").text.strip())
    except:
        pass
```

```python
prospect_info = pd.DataFrame({"player":prospects,"college":colleges,"position":positions,"height":heights,'weight':weights})
prospect_info
```

```python
prospect_info.to_csv("prospectinfo.csv", index=False)
```
